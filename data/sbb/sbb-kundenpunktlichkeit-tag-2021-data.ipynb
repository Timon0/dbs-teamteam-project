```python
import pandas as pd
import ssl

ssl._create_default_https_context = ssl._create_unverified_context
```

### Load Kundenpünktlichkeit file

```python
sbb_kundenpunktlichkeit_url = "https://data.sbb.ch/explore/dataset/kundenpunktlichkeit-tag/download/?format=csv&refine.datum=2021&timezone=Europe/Berlin&lang=de&use_labels_for_header=true&csv_separator=%3B"
sbb_kundenpunktlichkeit = pd.read_csv(sbb_kundenpunktlichkeit_url, sep=";")
sbb_kundenpunktlichkeit.head()
```

### Clean Data

```python
sbb_kundenpunktlichkeit.drop(columns=['Kundenpünktlichkeit KPP', 'KPP FV', 'KPP RV', 'KAP',
                                      'KAP FV', 'KAP FV', 'KAP RV', 'KCOR', 'cor', 'Schlüssel'], inplace=True)
sbb_kundenpunktlichkeit.head()
```

```python
sbb_kundenpunktlichkeit.columns = ['date', 'sbbregion_key', 'zugpuenktlichkeit', 'zugpuenktlichkeit_fv', 'zugpuenktlichkeit_rv']
```

```python
## convert date to expected format
def round_value(value):
    return round(value,10)
sbb_kundenpunktlichkeit['zugpuenktlichkeit']= sbb_kundenpunktlichkeit['zugpuenktlichkeit'].apply(round_value)
sbb_kundenpunktlichkeit['zugpuenktlichkeit_fv']= sbb_kundenpunktlichkeit['zugpuenktlichkeit_fv'].apply(round_value)
sbb_kundenpunktlichkeit['zugpuenktlichkeit_rv']= sbb_kundenpunktlichkeit['zugpuenktlichkeit_rv'].apply(round_value)
```

```python
sbb_kundenpunktlichkeit.sort_values(by=['date', 'sbbregion_key'] , inplace=True)
sbb_kundenpunktlichkeit.head()
```

### Export to csv

```python
sbb_kundenpunktlichkeit.to_csv( "sbb_kundenpunktlichkeit_data.csv", index=False, encoding='utf-8-sig', line_terminator='\n')
```
